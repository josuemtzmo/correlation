# Correlation coefficient between two series.

### Import libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
from correlation import Compute_Correlation
```

### Data example:

```python
x = np.linspace(0,2*np.pi,100)
y1 = 2*np.cos(x)
y2 = 2*np.sin(x)
```

```python
# Plot data
plt.plot(x,y1,x,y2)
```

Initiate python class:

```python
c_cor = Compute_Correlation(y1, y2,method="Cyclic")
```

Compute correlation coefficients

```python
corr,sig,t = c_cor.cor_series()
```

```python
# Plot correlation coefficient
plt.plot(corr,label="Correlation Coefficient")
# Plot Significance
plt.plot(corr*sig,color='steelblue',linewidth=8,alpha=0.5,label="Significance")
plt.legend()
plt.grid()
```

The correlation coefficients function aligns with the `sin` function, as `y1` is the signal leading in lag.

```python
ax = c_cor.plot()
ax.plot(corr,'-r',linewidth=3,label='Correlation')
plt.legend()
```

We can further visualise the correlation coefficient by animating the correlation for each lag:

```python
c_cor.animate()
```
